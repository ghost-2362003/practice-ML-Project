```python
#import scikit learn to create a custom dataset
from sklearn import datasets
import matplotlib.pyplot as plt   #for visualisation
import seaborn as sns
import pandas as pd   #for manipulating the dataframe
```

```python
# creating a dataset for classification using the .make_classification method of sklearn
X, _ = datasets.make_classification(2000, 7, n_clusters_per_class = 1, random_state = 3)
```

```python
X
```

```python
# visualizing the dataset at this point

plt.figure(figsize=(17,8))
plt.subplot(121, title='classification')
plt.scatter(X[:,0], X[:,1])
```

from the above visualisation we cannot understand how many clusters are there

```python
#applying k means clustering to understand the above data better
from sklearn.cluster import KMeans

pred = KMeans(n_clusters = 4, random_state = 45, verbose = 0).fit_predict(X)

plt.title('k-means clusters')
plt.scatter(X[:,0], X[:,1], c = pred)
```

The K-means algorithm did not do much of a work classifying the data points.
This goes to show that the data generated is not isotropic.
Also, the data is not well seperated. Thus, K-means has a hard time to assign each point the closest centroid

```python
import numpy as np
import math
from scipy.stats import norm

men_mean = 70
men_var = 1.5
men_std = math.sqrt(men_var)
men_dist = np.linspace(men_mean - 8 * men_std, men_mean + 8 * men_std, 4000)

women_mean = 67
women_var = 2
women_std = math.sqrt(women_var)
women_dist = np.linspace(women_mean - 8 * women_std, women_mean + 8 * women_std, 4000)

plt.figure(figsize=(17,8))
plt.title('Men\'s and Women\'s Height Distributions in Inches', fontdict={'fontsize':15})
plt.plot(men_dist, norm.pdf(men_dist, men_mean, men_std), 'b-', label='Men')
plt.plot(women_dist, norm.pdf(women_dist, women_mean, women_std), 'r-', label='Women')
plt.legend(fontsize=15)
```

This data is the height vs gender distributions between man and woman.
This data has the resemblance to a real life dataset with anisotropic variance
